Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

Entre -10 y 18 grados, que imprima "Fresquete".,
Entre 19 y 30 grados, que imprima "Calor'i".,
Entre 31 y 45 grados, que imprima "Hakuuuuuu".,
Para todo lo demás, imprimir "Ñamano mba'e!".

```python
temperatura = int(input("Ingresa la temperatura en grados Celsius"))

def grado(temperatura):
    if temperatura >-11 and temperatura <=18:
        print("Frequete")
    elif temperatura >18 and temperatura <=30:
        print("Calor'i")
    elif temperatura >30 and temperatura <=45:
        print("Hakuuuuuu")
    else:
        print("Ñamano mba'e!")

grado(temperatura)
print(f"La temperatura es {temperatura} grado")
```

Escribir un programa que almacene una contraseña en una variable, y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
contraseña = input("Ingresa la contraseña son 3 digitos numericos")
pass_correcta = "123"
while True:
    if contraseña == pass_correcta:
        break
    else:
        contraseña = input("Ingresa de vuelta la contraseña")
print("Contraseña es correcta")
print("Persevera y triunfaras") 
```
