Crear función que revise si una palabra contiene una letra. En caso de que la contenga, retornar la letra entre corchetes []. De lo contrario, retornar solo la letra.

Definir variable letra = input("Ingresa una palabra")
Ver si la palabra tiene una letra especifica
Definir una funcion def verificardor()
contiene es in
realizar un return en corchetes si tiene la letra clave
si no contiene la letra especifica que traiga solo la letra. 

```python
palabra_ingresada = input("Ingrese una palabra")
condicion = "a"
def verificador():
    for p_ingresada in palabra_ingresada:
        if p_ingresada == condicion:
            return [condicion]
        else:
            return condicion
        
verificador()
```

Pedir al usuario que ingrese 1 palabra, luego convertir esa palabra a una lista con sus caracteres con list(). Agregar esa lista de caracteres a una lista llamada palabras_ingresadas. Repetir el proceso 3 veces. Cada vez que se agregue una lista de caracteres, imprimir la lista palabras_ingresadas.

crear una variable input
convertir la palabra de la variable en list()
agregar esa lista en palabras_ingresadas.
////////////////////////////////////////////////////

```python
palabras_ingresadas = []
for i in range(3):
    palabra = input("Ingresa una palabra aleatoria")
    lista_palabra = list(palabra)
    palabras_ingresadas.append(lista_palabra)
    print(palabras_ingresadas)
```
