Suma: asignar dos números en dos variables diferentes cantidad_de_manzanas y cantidad_de_naranjas. Luego sumar ambas variables en cantidad_de_frutas. Imprimir la última variable.

```python
cantidad_de_manzanas =  5
cantidad_de_naranjas = 5
cantidad_de_frutas = cantidad_de_manzanas + cantidad_de_naranjas
print("La totalidad de frutas son", cantidad_de_frutas)
```

Resta: asignar un valor numérico a la variable frutas_vendidas. Luego, asignar la resta entre cantidad_de_frutas y frutas_vendidas a la variable frutas_restantes. Imprimir la última variable.

```python
frutas_vendidas =  4
frutas_restantes = cantidad_de_frutas - frutas_vendidas
print("Las frutas restantes son" , frutas_restantes)
```

Además, deben trabajar con variables locales y globales.

```python
# Variable global cuando la variable esta fuera de la funcion
color = input("Ingresa el color de la motocicleta")

def caracteristicas(caracteristica1, caracteristica2):
    # Variable local cuando la variable esta dentro de la funcion
    marca = input("Ingresa la marca de la motocicleta")
    print(f"Que caracteristica tiene esta motocicleta?. Esta motocicleta es de color {color} y de la marca {marca}")
    
caracteristicas (caracteristica1, caracteristica2)
```
